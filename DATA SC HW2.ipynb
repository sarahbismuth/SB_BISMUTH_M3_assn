```python
#i decided to replace all the missing values by the mean or the mode  except for the STAR column , since they are continuous values
import numpy as np
gg['Chlorides'].mean()
gg['Chlorides'].replace(np.NaN,gg['Chlorides'].mean(),inplace=True)

gg['pH'].mode()
gg['pH'].replace(np.NaN,gg['Chlorides'].mode(),inplace=True)

gg['FreeSulfurDioxide'].mean()
gg['FreeSulfurDioxide'].replace(np.NaN,gg['FreeSulfurDioxide'].mean(),inplace=True)

gg['Alcohol'].mean()
gg['Alcohol'].replace(np.NaN,gg['Alcohol'].mean(),inplace=True)

gg['TotalSulfurDioxide'].mean()
gg['TotalSulfurDioxide'].replace(np.NaN,gg['TotalSulfurDioxide'].mean(),inplace=True)


gg['Sulphates'].mean()
gg['Sulphates'].replace(np.NaN,gg['Sulphates'].mean(),inplace=True)

gg['ResidualSugar'].mean()
gg['ResidualSugar'].replace(np.NaN,gg['ResidualSugar'].mean(),inplace=True)
```
